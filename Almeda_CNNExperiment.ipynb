```python
import numpy as np
import cv2
import os 
import random
import matplotlib.pyplot as plt
import pickle
```

```python
DIRECTORY = "dataset"
CATEGORIES = ["cat", "dog"]
```

```python
for category in CATEGORIES:
    datasets = os.path.join(DIRECTORY, category)
    print(datasets)
    for i, img in enumerate(os.listdir(datasets)):
        img_path = os.path.join(datasets, img)
        if i % 100 == 0:
            print(img_path)
```

```python
IMG_SIZE = 64

for category in CATEGORIES:
    datasets = os.path.join(DIRECTORY, category)
    for img in os.listdir(datasets):
        img_path = os.path.join(datasets, img)
        img_arr = cv2.imread(img_path, 0)
        img_arr = cv2.resize(img_arr,(IMG_SIZE, IMG_SIZE))
        plt.imshow(img_arr, cmap='gray')
        break
```

```python
IMG_SIZE = 64
data = []
for category in CATEGORIES:
    datasets = os.path.join(DIRECTORY, category)
    label = CATEGORIES.index(category)
    for img in os.listdir(datasets):
        img_path = os.path.join(datasets, img)
        img_arr = cv2.imread(img_path)
        img_arr = cv2.resize(img_arr,(IMG_SIZE, IMG_SIZE))
        data.append([img_arr, label])
```

```python
len(data)
```

```python
random.shuffle(data)
```

```python
X = []
y = []

for features, labels in data:
    X.append(features)
    y.append(labels)
```

```python
y
```

```python
X = np.array(X)
y = np.array(y)
```

```python
pickle.dump(X, open('X.pkl', 'wb'))
pickle.dump(y, open('y.pkl', 'wb'))
```

```python
X = pickle.load(open('X.pkl', 'rb'))
y = pickle.load(open('y.pkl', 'rb'))
```

```python
X = X/255
X.shape
```

```python
VAL_PCT = 0.3
val_size = int(len(X)*VAL_PCT)
val_size
```

```python
train_X = X[: -val_size]
train_y = y[: -val_size]

test_X = X[-val_size:]
test_y = y[-val_size:]

print(train_X.shape)
print(test_X.shape)

split_set = train_X, test_X, train_y, test_y
```

```python
import tensorflow as tf
#creates models
def create_CNNModel(hl, ol, act):

    cnn = tf.keras.models.Sequential()
    cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation=hl, input_shape=[64, 64, 3]))
    cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))

    cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation=hl))
    cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))

    cnn.add(tf.keras.layers.Flatten())

    cnn.add(tf.keras.layers.Dense(units=128, activation=hl))

    cnn.add(tf.keras.layers.Dense(units=1, activation=ol))

    cnn.compile(optimizer=act, loss = 'binary_crossentropy', metrics = ['accuracy'])

    return cnn  

import pandas as pd
from collections import defaultdict
hl_f = ['sigmoid','tanh','relu','LeakyReLU','PReLU']
ol_f = ['sigmoid','tanh','relu','LeakyReLU','PReLU']
act = ['SGD', 'RMSprop','Adam','Adagrad','Adamax']

table = defaultdict(list)

for i in hl_f:
    for j in ol_f:
        for k in act:
            table['hl_f'].append(i)
            table['ol_f'].append(j)
            table['k_f'].append(k)

            model = create_CNNModel(i, j, k)
            model.fit(train_X, train_y, batch_size=32, epochs=25)
            score = model.evaluate(test_X, test_y, batch_size=32)
            table['score'].append(score[1])
            with open('accuracies.txt', 'a') as f:
                f.write('\n'+str(score[1]))
```

```python
df = pd.DataFrame(table)
```

```python
df.to_csv('ExperimentTable.csv')
```
